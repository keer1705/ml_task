```python
import pandas as pd
```

```python
df1=pd.read_csv('./boston_house/HousingData.csv')
```

```python
print("Null Values:")
print(df1.isnull().sum())
```

```python
df1_nulval=df1.dropna()
print(df1_nulval.isnull().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
numco= df1.select_dtypes(include='number').columns
plt.figure(figsize=(15, 8))
df1[numco].boxplot()
plt.title("Visualize Outliers using Boxplot")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
numm=df1[numco]
scaler=StandardScaler()
y_pred=scaler.fit_transform(numm)
dfstd = pd.DataFrame(y_pred, columns=numco)
print("Standard Scaler Summary",dfstd.describe())
```

```python
mmsc=MinMaxScaler()
y_pred=mmsc.fit_transform(numm)
df_mmsc = pd.DataFrame(y_pred, columns=numco)
print("Min Max Scaler Summary",df_mmsc.describe())
```
