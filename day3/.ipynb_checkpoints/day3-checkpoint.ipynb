```python
# Plot a histogram for sepal length
# Plot sepal length vs sepal width as a scatter plot
# Generate a pairplot for the dataset
# Draw a boxplot for petal width grouped by species
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../iris/iris.csv')
print(df.head())
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['SepalLengthCm'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram for Sepal Length')
plt.xlabel('Sepal Length cm')
plt.ylabel('Sepal Width cm')
plt.show()
```

```python
plt.scatter(df['SepalLengthCm'],df['SepalWidthCm'])
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
df.boxplot(column='PetalWidthCm', by='Species')
plt.title('BoxPlot for PetalWidthCm grouped with Species')
plt.xlabel('Species')
plt.ylabel('PEtal Width')
plt.show()
```
